```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:\\Users\\cheta\\Downloads\\iris.csv")
```

```python
data
```

```python
data.dtypes
```

```python
plt.hist(data['SepalLengthCm'],bins=10, edgecolor='g', alpha=0.7)
plt.xlabel('Sepal Length Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
plt.hist(data['SepalWidthCm'],bins=10, edgecolor='g', alpha=0.7)
plt.xlabel('Sepal Width Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')
plt.show()
```

```python
plt.hist(data['PetalLengthCm'],bins=10, edgecolor='g', alpha=0.7)
plt.xlabel('petal Length Cm')
plt.ylabel('Frequency')
plt.title('Histogram of petal Length')
plt.show()
```

```python
plt.hist(data['PetalWidthCm'],bins=10, edgecolor='g', alpha=0.7)
plt.xlabel('petal Width Cm')
plt.ylabel('Frequency')
plt.title('Histogram of petal Width')
plt.show()
```

```python
species_count = data['Species'].value_counts()
plt.bar(species_count.index, species_count.values)
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.title('Distribution of Species')
plt.show
```
